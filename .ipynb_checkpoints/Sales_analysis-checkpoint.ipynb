```python
# merge sales data 
# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
import os
```

```python
files = [file for file in os.listdir('./Sales_Data')]
for file in files:
    print(file)
```

```python
#concat files
all_months_data = pd.DataFrame() #create empty data frame
for file in files:
    df = pd.read_csv("./Sales_Data/"+file) #read all dataframes
    all_months_data = pd.concat([all_months_data, df]) #concating data

all_months_data.head()
```

```python
#export new data to csv to check whether the data is combined
all_months_data.to_csv("all_data.csv", index = False)
```

```python
#read in new file
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
# additional column for "month"
all_data['Month'] = all_data['Order Date'].str[0:2] #take first two strings of the order date as month column
```

```python
all_data.head()
```

```python
#clean data NaN
# https://stackoverflow.com/questions/43424199/display-rows-with-one-or-more-nan-values-in-pandas-dataframe
nan_df = all_data[all_data.isna().any(axis=1)] # filter null fields
nan_df.head()
```

```python
#drop empty rows
all_data = all_data.dropna(how='all') #dropping all empty data 
all_data.head()
```

```python
#find and drop all 'Or'
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or'] #find all raws with 'Or'
temp_df.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or'] #filter out 'Or' values
all_data.head()
```

```python
all_data['Month'] = all_data['Month'].astype('int32') #converting str to int
all_data.head()
```

```python
#convert column data types

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #str to int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #str to float
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each'] #add a calculated sales column
all_data.head()
```

```python
all_data.sort_values(by=['Sales'], inplace=True, ascending=False) # order by sales
all_data.head()
```

### the best month for sales

```python
yearlysales = all_data.groupby('Month').sum()
yearlysales
# all_data.groupby('Month').sum()['Sales'] # filtered by sales values only
```

```python
#visualize sales

import matplotlib.pyplot as plt # import visoalization


months = range(1,13) # adding the momnth range

plt.bar(months, yearlysales['Sales']) #filter the plotting to sales
plt.xticks(months) #visualizes all the months 
plt.ylabel('Sales in USD ($)') # y axis
plt.xlabel('Month') # xaxis
plt.show()
```

### city with the highest number of sales

```python
# all_data['City'] =all_data['Purchase Address'].apply(lambda x: x.split(',')[1]) #for each of the cell value, split by comma and get the first index


def get_city(address):
    return address.split(',')[1] #for each of the cell value, split by comma and get the first index


def get_state(address):
    return address.split(',')[2].split(' ')[1] #for each of the cell value, split by comma and get the second, after the white space we get the first index which is the country code 

# all_data['State'] =all_data['Purchase Address'].apply(lambda x: get_state(x)


#creating a new cell with the city and state
all_data['City'] =all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x)+ ')')                                                  
```

```python
all_data.head()
```

```python
city_sales = all_data.groupby('City').sum() #sale grouped by cities
```

```python
city_sales.head()
```

```python
#visualize sales data per city

# cities = all_data['City'].unique() # returns the cities in a wrong order while plotting and showing inconsistency 

cities = [city for city, df in all_data.groupby('City')] # align x and y data by looping through the cities to have them in the same order as the grouped sales data
plt.bar(cities, city_sales['Sales']) #filter the plotting to sales and cities
plt.xticks(cities, rotation='vertical', size=8) #visualizes all the cities 
plt.ylabel('Sales in USD ($)') # y axis
plt.xlabel('City Name') # xaxis
plt.show()
```
