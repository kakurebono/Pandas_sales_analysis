```python
# merge sales data 
# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
import os
```

```python
files = [file for file in os.listdir('./Sales_Data')]
for file in files:
    print(file)
```

```python
#concat files
all_months_data = pd.DataFrame() #create empty data frame
for file in files:
    df = pd.read_csv("./Sales_Data/"+file) #read all dataframes
    all_months_data = pd.concat([all_months_data, df]) #concating data

all_months_data.head()
```

```python
#export new data to csv to check whether the data is combined
all_months_data.to_csv("all_data.csv", index = False)
```

```python
#read in new file
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
# additional column for "month"
all_data['Month'] = all_data['Order Date'].str[0:2] #take first two strings of the order date as month column
```

```python
all_data.head()
```

```python
#clean data NaN
# https://stackoverflow.com/questions/43424199/display-rows-with-one-or-more-nan-values-in-pandas-dataframe
nan_df = all_data[all_data.isna().any(axis=1)] # filter null fields
nan_df.head()
```

```python
#drop empty rows
all_data = all_data.dropna(how='all') #dropping all empty data 
all_data.head()
```

```python
#find and drop all 'Or'
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or'] #find all raws with 'Or'
temp_df.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or'] #filter out 'Or' values
all_data.head()
```

```python
all_data['Month'] = all_data['Month'].astype('int32') #converting str to int
all_data.head()
```

```python
#convert column data types

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #str to int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #str to float
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each'] #add a sales column
all_data.head()
```

```python
all_data.sort_values(by=['Sales'], inplace=True, ascending=False) # order by sales
all_data.head()
```

```python
# find the best month for sales
all_data.groupby('Month').sum().reset_index()
all_data.sort_values('Month', ascending=False)
all_data.head()

# grouped = df.groupby('mygroups').sum().reset_index()
# grouped.sort_values('mygroups', ascending=False)
```
